```python
import cv2
from PIL import Image
import numpy as np
import re
```

```python
# 图片二值化
from matplotlib.pyplot import axis


img = Image.open('map.png')

ROTATE=True
#ROTATE=False

if(ROTATE):
    img= img.transpose(Image.ROTATE_270)
    img=img.resize((122,250))
else:
    img=img.resize((122,250))



#img.show()
 
# 模式L”为灰色图像，它的每个像素用8个bit表示，0表示黑，255表示白，其他数字表示不同的灰度。
Img = img.convert('L')
Img.save("gray.jpg")
 
# 自定义灰度界限，大于这个值为黑色，小于这个值为白色
threshold = 200
 
table = []
for i in range(256):
    if i < threshold:
        table.append(0)
    else:
        table.append(1)
 
# 图片二值化
photo = Img.point(table, '1')

#photo.show()
photo.save("bw.jpg")

expan_arr=np.zeros((250,6))
data=np.array(photo).astype(int)

print(data.shape)
print(expan_arr.shape)
print('part a',data[:,:120].shape)
print('part c',data[:,120:].shape)
data=np.concatenate((data,expan_arr),axis=1).astype(int)
width,height=data.shape
#data=data.reshape(((width*height),1)).astype(int)



#250,128
data_hex=[]
sum=0

'''for i in range(int(data.size/8)):
    sum=0
    for j in range(8):
        eleAtPos=data[(8*i+j),0]
        sum=sum+(eleAtPos*(2**j))
    data_hex.append("{:#04X}".format(sum))'''

for i in range(data.shape[0]):
    for j in range(data.shape[1]//8):
        sum=0
        for r in range(8):
            eleAtPos=data[i,8*j+r]
            sum=sum+(eleAtPos*(2**(7-r)))
        data_hex.append("{:#04X}".format(sum))
            


file = open('map.txt','w')

file.write(str(data_hex))

file.close()


f=open('map.txt','r')
alllines=f.readlines()
f.close()
f=open('map.txt','w+')
for eachline in alllines:
    a=re.sub('\'','',eachline)
    a=re.sub('\[','',a)
    a=re.sub('\]','',a)
    #a=re.sub(']','',a)
    f.writelines(a)
f.close()
```
